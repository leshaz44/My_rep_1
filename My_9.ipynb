Напишите функцию, которая принимает строку и возвращает True, если строка является палиндромом (игнорируя регистр и пробелы), и False в противном случае.

Напишите функцию, которая принимает строку и возвращает True, если строка является палиндромом (игнорируя регистр и пробелы), и False в противном случае.

```python
def is_palindrome(s: str) -> bool:
    # Удаляем небуквенно-цифровые символы и приводим строку к нижнему регистру
    s_cleaned = "".join(filter(str.isalnum, s)).lower()
    # Сравниваем очищенную строку с ее перевернутой версией
    return s_cleaned == s_cleaned[::-1]

# Примеры использования:
print(f"'А роза упала на лапу Азора' является палиндромом: {is_palindrome('А роза упала на лапу Азора')}")
print(f"'Madam, I\'m Adam' является палиндромом: {is_palindrome('Madam, I\'m Adam')}")
print(f"'Привет' является палиндромом: {is_palindrome('Привет')}")
print(f"'Racecar' является палиндромом: {is_palindrome('Racecar')}")
print(f"'A man, a plan, a canal: Panama' является палиндромом: {is_palindrome('A man, a plan, a canal: Panama')}")
```
